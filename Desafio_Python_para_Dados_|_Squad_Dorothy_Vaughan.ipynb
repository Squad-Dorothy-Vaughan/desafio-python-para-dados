## Pergunta 02
Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mode]

````python
# lógica aqui
# groupby -> cria uma cópia do dataframe automaticamente.

df_moda_mediana_media_sono = df.groupby('Profissão')['Duração do sono'].agg(
    Media='mean',
    Mediana=np.median,
    Moda=lambda m: pd.Series.mode(m).iloc[0] if not pd.Series.mode(m).empty else np.nan
).reset_index()

df_moda_mediana_media_sono
````

#### Explicação geral da lógica (escreva aqui, a lógica usada para o código desenvolvido abaixo):

A partir do dataframe df que temos, criamos um novo dataframe chamado df_moda_mediana_media_sono com as colunas Profissão, Media, Mediana, Moda. Para conseguir realizar os cálculos e pegar as devidas colunas do df original, usamos o groupby(), método do pandas usado para dividir os dados de um DataFrame em grupos com base em uma ou mais colunas.

No nosso caso, pegamos a coluna 'Profissão' para que, cada registro de profissão seja uma linha. Já a coluna 'Duração do sono' é o parâmetro usado para definir qual a informação que vamos utilizar para realizar os cáculos de média, moda e mediana.

Logo depois, usamos o .agg() para realizar as 3 operações nas horas de sono. dentro dos () do método .agg criamos as colunas chamadas Media, Mediana e Moda para que em cada uma, possamos ter os resultados a partir dos cálculos feitos com as horas de sono.

Coluna Media: criamos a coluna Media e passamos o parâmetro ='mean' para que ela realize uma média aritmética simples (somar cada hora de sono de cada respondente que tem x profissão e dividir pela quantidade de respondentes da profissão).

Coluna Mediana: utilizamos a biblioteca numpy (np) para calcular a mediana e achar no raio de valores de duração de sono o valor que está exatamente no meio (isso é feito por profissão).
EX: [1, 2, 3, 4, 5] - RETURN: 3

Coluna Moda: para descobrimos qual valor de duração de sono mais se repete em cada profissão, calculamos a moda.

Função lambda: O Pandas pode retornar mais de uma moda se houver empate (ex: se 7 horas e 8 horas se repetirem a mesma quantidade de vezes), a linha com lambda (função que permite a criação de outras funções sem a necessidade de usar a estrutura tradicional). Na linha lambda, nós temos:

* Calculo da moda (pd.Series.mode(m)).
* Verificação se o resultado não está vazio (if not ... empty).
* Pegando apenas o primeiro valor encontrado (.iloc[0]) para não quebrar a tabela. Se estiver vazio, joga np.nan (valor nulo).

.iloc[0]: Garante que, independentemente de haver empate ou não na moda, o resultado final devolvido para a tabela seja sempre um número único e puro, mantendo a estrutura idêntica às colunas de Média e Mediana.

Finalizando com .reset_index():
Por padrão, quando usamos o groupby, a coluna usada para agrupar (Profissão) vira o "índice" (o identificador das linhas) da tabela. O método .reset_index() reseta isso e transforma 'Profissão' em uma coluna comum. Isso deixa o DataFrame final formatado como uma tabela padrão, com as colunas: Profissão, Media, Mediana e Moda


## Pergunta 03
Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?


````python
# Filtrando o grupo
engenharia = df[df['Profissão'] == 'Eng. de Software']

# Contando as linhas
total_engenharia = len(engenharia)
obesos_engenharia = len(engenharia[engenharia['Categoria IMC'] == 'Obesidade'])

# Calculando a porcentagem
porcentagem_obesos = (obesos_engenharia / total_engenharia) * 100
print(f"Total de profissionais em Engenharia de Software: {total_engenharia}")
print(f"Total de engenheiros na categoria Obesidade: {obesos_engenharia}")
print(f"Porcentagem de obesos no grupo: {porcentagem_obesos:.2f}%")
````

#### Explicação geral da lógica (escreva aqui, a lógica usada para o código desenvolvido acima):

Para resolver a questão 3, a lógica foi isolar primeiro os dados do grupo de interesse para não misturar os resultados. Para isso, foi feito um filtro na coluna de profissão usando o termo exato 'Eng. de Software'. Em seguida, utilizamos o comando len para fazer duas contagens: o total de profissionais dessa área e quantos deles tinham a indicação de 'Obesidade' na coluna de IMC. O passo final foi calcular a percentagem dividindo a quantidade de obesos pelo total de engenheiros e multiplicando por 100. Como a base de dados possui apenas 4 engenheiros de software e só 1 está nessa categoria, o resultado final foi de 25%.

## Pergunta 04
De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)

````python
# Filtrar as duas profissões simultaneamente usando o método .isin()
````

````python
# Filtrando as duas profissões usando o método .isin()
advogados_vendas = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]

# Média de sono do grupo vs Média geral de toda a base
media_grupo = advogados_vendas.groupby('Profissão')['Duração do sono'].mean()
media_geral = df['Duração do sono'].mean()

print(f"Média geral de sono da base: {media_geral:.2f} horas\n")
print("Média de sono das profissões analisadas:")
print(media_grupo)
````

#### Explicação geral da lógica (escreva aqui, a lógica usada para o código desenvolvido acima):

Para responder à questão 4, utilizei o método .isin() para fazer uma filtragem dupla na coluna de profissão, o que permitiu isolar os dados de Advogados e Representantes de Vendas ao mesmo tempo. Com esse subconjunto criado, apliquei o groupby combinado com o mean para calcular a média específica de horas de sono de cada uma dessas duas funções. Para servir de parâmetro de comparação, também calculei a média de sono da base de dados completa, que resultou em 7.13 horas. Ao cruzar as informações, ficou evidente que os advogados dormem uma média de 7.41 horas, ficando acima do índice geral, enquanto os representantes de vendas têm uma média de apenas 5.90 horas. Portanto, conclui-se que apenas a profissão de Representante de Vendas faz a pessoa dormir menos do que a média geral do grupo analisado.

## Pergunta 05
Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)

````python
# Pergunta 05
# Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono?

# Filtrando apenas as profissões Médico(a) e Enfermeiro(a)
profissoes_saude = df[df["Profissão"].isin(["Médico(a)", "Enfermeiro(a)"])]

# Calculando a média de duração do sono para cada profissão
media_sono_saude = profissoes_saude.groupby("Profissão")["Duração do sono"].mean()

# Exibindo as médias
print(media_sono_saude)

# Identificando quem tem a menor média
menor_media = media_sono_saude.idxmin()
valor_menor_media = media_sono_saude.min()

# Exibindo a resposta final
print(f"\nEntre Enfermeiro(a) e Médico(a), quem tem menor média de horas de sono é {menor_media}, com {valor_menor_media:.2f} horas.")
````

#### Explicação geral da lógica (escreva aqui, a lógica usada para o código desenvolvido acima):

Para responder essa pergunta, primeiro filtramos o DataFrame utilizando o método isin(), selecionando apenas os registros em que a profissão era Médico(a) ou Enfermeiro(a). Em seguida, agrupamos os dados pela coluna Profissão com groupby() e calculamos a média da coluna Duração do sono com mean(). Depois, usamos idxmin() para identificar qual profissão teve a menor média e min() para obter o valor dessa menor média. O resultado mostra que Médico(a) apresenta menor média de horas de sono, com aproximadamente 6,97 horas.

## Pergunta 07

Por fim, reestruturei o nome das colunas para 'Profissão' e 'Quantidade' e utilizei o .iloc[[-1]] para isolar o último registro da base em formato de tabela, que representa justamente a profissão com a menor recorrência entre os entrevistados.

## Pergunta 10

Em seguida, utilizei a coluna 'Passos diários' e o método .mean() para calcular a média de passos de cada condição. Para exibir os resultados sem casas decimais, utilizei a função int().

A média de passos das pessoas com frequência cardíaca acima de 70 foi de 6472 passos, enquanto a média das pessoas com frequência cardíaca menor ou igual a 70 foi de 7006 passos.

Portanto, as pessoas com frequência cardíaca menor ou igual a 70 deram em média mais passos.
